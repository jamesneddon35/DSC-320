```python
housing_df = pd.read_csv('/Users/jamessneddon/Downloads/week1data/housing_data.csv')
```

```python
housing_df.head()
```

```python
y_true = housing_df['sale_price'].values
y_pred = housing_df['sale_price_pred'].values
#convert variables to NumPy Arrays
```

```python
rmse = np.sqrt((1 / y_true.size) * np.sum((y_true - y_pred) ** 2))
print("RMSE:", rmse)
#apply formula using numpy
```

```python
mae = np.sqrt((1 / y_true.size) * np.abs(np.sum(y_true - y_pred)))
print("MAE:", mae)
#ditto
```
